## 2.1 Genertate indicators
### Overlap Studies Functions 重叠研究指标
#### BBANDS - Bollinger Bands
函数名：BBANDS
名称： 布林线指标
简介：其利用统计原理，求出股价的标准差及其信赖区间，从而确定股价的波动范围及未来走势，利用波带显示股价的安全高低价位，因而也被称为布林带。
#### DEMA - Double Exponential Moving Average 双移动平均线

函数名：DEMA
名称： 双移动平均线
简介：两条移动平均线来产生趋势信号，较长期者用来识别趋势，较短期者用来选择时机。正是两条平均线及价格三者的相互作用，才共同产生了趋势信号。
#### EMA - Exponential Moving Average

函数名：EMA
名称： 指数平均数
简介：是一种趋向类指标，其构造原理是仍然对价格收盘价进行算术平均，并根据计算结果来进行分析，用于判断价格未来走势的变动趋势。
#### KAMA - Kaufman Adaptive Moving Average 考夫曼的自适应移动平均线

函数名：KAMA
名称： 考夫曼的自适应移动平均线
简介：短期均线贴近价格走势，灵敏度高，但会有很多噪声，产生虚假信号；长期均线在判断趋势上一般比较准确 ，但是长期均线有着严重滞后的问题。我们想得到这样的均线，当价格沿一个方向快速移动时，短期的移动 平均线是最合适的；当价格在横盘的过程中，长期移动平均线是合适的。

#### MA - Moving average 移动平均线

函数名：MA
名称： 移动平均线
简介：移动平均线，Moving Average，简称MA，原本的意思是移动平均，由于我们将其制作成线形，所以一般称之为移动平均线，简称均线。它是将某一段时间的收盘价之和除以该周期。 比如日线MA5指5天内的收盘价除以5 。

#### MAMA - MESA Adaptive Moving Average

#### SMA - Simple Moving Average 简单移动平均线

函数名：SMA
名称： 简单移动平均线
简介：移动平均线，Moving Average，简称MA，原本的意思是移动平均，由于我们将其制作成线形，所以一般称之为移动平均线，简称均线。它是将某一段时间的收盘价之和除以该周期。 比如日线MA5指5天内的收盘价除以5。

#### TEMA - Triple Exponential Moving Average

函数名：TEMA（T3 区别？） 名称：三重指数移动平均线

#### TRIMA - Triangular Moving Average

#### WMA - Weighted Moving Average 移动加权平均法

函数名：WMA
名称：加权移动平均线
简介：移动加权平均法是指以每次进货的成本加上原有库存存货的成本，除以每次进货数量与原有库存存货的数量之和，据以计算加权平均单位成本，以此为基础计算当月发出存货的成本和期末存货的成本的一种方法。
#### T3 - Triple Exponential Moving Average (T3) 三重指数移动平均线

函数名：T3
名称：三重指数移动平均线
简介：TRIX长线操作时采用本指标的讯号，长时间按照本指标讯号交易，获利百分比大于损失百分比，利润相当可观。 比如日线MA5指5天内的收盘价除以5 。

```python
def generate_indicators(df, start_date, end_date):
    indicators=pd.DataFrame(df)
    
    #Overlap Studies Functions 重叠研究指标
    MA_types=['SMA','EMA','WMA','DEMA','TEMA', 'TRIMA','KAMA','MAMA','T3']
    for i in range(len(MA_types)):
        indicators[MA_types[i]] = ta.MA(df.Close,timeperiod=5,matype=i)
    
    #Different timeperiod EMA
    N=[5,10,30,120]
    for i in N:
        indicators['ma_'+str(i)] = ta.EMA(df.Close,timeperiod=i)
    #BBANDS - Bollinger Band
    H_line,M_line,L_line = ta.BBANDS(df.Close, timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
    indicators['H_line'] = H_line
    indicators['M_line'] = M_line
    indicators['L_line'] = L_line
    #Others
    indicators['HT']=ta.HT_TRENDLINE(df.Close)
    periods =np.array([3]*len(df), dtype=float)
    indicators['MAVP']=ta. MAVP(df.Close,periods)
    indicators['MIDPOINT']=ta.MIDPOINT(df.Close)
    indicators['MIDPRICE']=ta.MIDPRICE(df.High,df.Low)
    indicators['SAR']=ta.SAR(df.High,df.Low)
    indicators['SAREXT']=ta.SAREXT(df.High,df.Low)
    
    
    #Momentum Indicator Functions 动量指标
    #ADX - Average Directional Movement Index
    indicators['ADX'] = ta.ADX(df.High, df.Low, df.Close, timeperiod=14)
    #APO - Absolute Price Oscillator
    indicators['APO'] = ta.APO(df.Close, fastperiod=12, slowperiod=26, matype=0)
    #CCI
    indicators['CCI'] = ta.CCI(df.High, df.Low, df.Close, timeperiod=14)
    #MACD - Moving Average Convergence/Divergence
    indicators['MACD_blue'], indicators['MACD_red'], indicators['MACD_hist'] = ta.MACD(df.Close, fastperiod=12, slowperiod=26, signalperiod=9)
    #CMO - Chande Momentum Oscillator
    indicators['CMO'] = ta.CMO(df.Close, timeperiod=14)
    #TRIX
    indicators['TRIX'] = ta.TRIX(df.Close, timeperiod=30)
    #RSI
    indicators['RSI'] = ta.RSI(df.Close, timeperiod=14)
    
    
    #Statistic Functions 统计学指标
    #BETA
    indicators['BETA'] = ta.BETA(df.High, df.Low, timeperiod=5)
    #CORREL - Pearson's Correlation Coefficient (r)
    indicators['CORREL'] = ta.CORREL(df.High, df.Low, timeperiod=30)
      

    return indicators
```

## 2.3 Price Return

```python
def cal_log_return(data):
    result = []
    for i in range(data.shape[0]-1):
        result.append(log(data.iloc[i+1])-log(data.iloc[i]))
    result = np.array(result)
    return result
```

```python
def cal_return(data):
    result = []
    for i in range(data.shape[0]-1):
        if data.iloc[i] == 0:
            print('0000000000000000000000000')
        result.append((data.iloc[i+1]-data.iloc[i])/data.iloc[i])
    result = np.array(result)
    return result
```

## 2.4 Wavelet

```python
def generate_wavelet(x, dt):
    # and a sample spacing
    dt = dt
    wa = WaveletAnalysis(x, dt=dt)
    # wavelet power spectrum
    power = wa.wavelet_power
    # scales 
    scales = wa.scales
    # associated time vector
    t = wa.time
    # reconstruction of the original data
    rx = wa.reconstruction()
    # plot_wavelet(t, scales, power)
    return rx,t,power,scales
```

```python
def plot_compare(data, return_format = 'log'):
    if return_format == 'log':
        s1= data
        s2 = cal_log_return(data)
        rx,t,power,scales = generate_wavelet(np.array(cal_log_return(data)), 1)
    elif return_format == 'normal':
        s1= data
        s2 = cal_return(data)
        rx,t,power,scales = generate_wavelet(np.array(cal_return(data)), 1)

    plt.style.use('default')

#     plt.style.use('ggplot')
    fig= plt.figure(figsize=[6,7],dpi=200)
       

#     fig= plt.figure()
#     fig.set_size_inches(6, 5, forward=True)
    fig.suptitle(' Wavelet transform for intra-day data [S&P500 index 2014-04-11]', y=1)
    # axs1=plt.subplot(311,facecolor='#FFFFFF')
    axs1=plt.subplot(311)
    axs1.set_title('S&P500 price')
    axs1.set_xlabel('Time')
    axs1.set_ylabel('Price')
    axs1.grid(True)
    axs1.plot(s1)

#     axs2=plt.subplot(312,facecolor='#FFFFFF')
    axs2=plt.subplot(312)
    axs2.set_title('S&P500 return')
    axs2.set_xlabel('Time')
    axs2.set_ylabel('Return')
    axs2.axis(xmin=0,xmax=388)
    axs2.grid(True)
    axs2.plot(s2)       

    axs3=plt.subplot(313) 
    axs3.set_title('Power Spectrum in Differen Scales')
    axs3.set_xlabel('Time')
    axs3.set_ylabel('Power')
    axs3.axis(xmin=0,xmax=388)
    axs3.grid(True)
    # axs3.plot(power)
    for i in range(power.shape[0]):
        axs3.plot(power[i,:])
    # axs3.legend(title='Parameter where:')
    fig.tight_layout()
    fig.savefig('figure/wavelet_compare.png',overwrite=True)

```

```python
 def plot_wavelet_color(t, scales, power):
    fig, ax = plt.subplots()
    T, S = np.meshgrid(t, scales)
    # ax.matshow(scales)
    ax.set_title('wavelet_power_spectrum')
    ax.set_xlabel('word length anomaly')
    ax.set_ylabel('sentence length anomaly')
    ax.contourf(T, S, power, levels=1000)
    ax.set_yscale('log')
    ax.plot()
    fig.savefig('figure/wavelet_power_spectrum_2d.png')
```

```python
def plot_wavelet_bw(t, scales, power):
    fig, ax = plt.subplots(dpi=200)
    ax.grid(False)
    scaler = MinMaxScaler( feature_range=(0, 1))
    power_m = scaler.fit_transform(power)
    # ax.set_title('wavelet_power_spectrum')
    ax.set_xlabel('time')
    ax.set_ylabel('scale')

    ax.imshow(power_m, cmap='gray', vmin=0, vmax=1)
    fig.savefig('figure/input_2d.png')
```

```python
def plot_wavelet_3D(t, scales, power):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    Y, X = np.meshgrid(t, scales)
    Z = power
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False)
#     ax.set_zscale('log')
    ax.set_ylabel('fourier period (minutes)')#
    ax.set_xlabel('Time')
    fig.colorbar(surf, shrink=0.5, aspect=5)
    fig.tight_layout()
    plt.show()
    fig.savefig('figure/wavelet_power_spectrum_3D.png')
```

# 3. Generate training and testing data

```python
def generate_train_data(data, start_date, end_date, full_day, indicators_index='Close'):
    
    date_time = []
    for date_i in date_range(start_date,end_date):
        date_i = date_i.isoformat()
        if date_i in full_day:
            date_time.append(date_i)
    # print(date_time)
    num = len(date_time)
    row = 61
    col = 360
    x_train = np.zeros([num,row,col,len(indicators_index)],dtype = np.float32)
    y_train = np.zeros([num],dtype = np.float32)
    count = 0
    sc = MinMaxScaler(feature_range=(0,1))
    for date_i in tqdm(date_time):
        count1=0
        for j in range(len(indicators_index)):
            temp_data = data[date_i][indicators_index[j]][datetime.time(8,30):datetime.time(14,30)]
            if indicators_index[j] in ['Open','High','Low','Close','SMA','EMA','WMA','DEMA','TEMA', 'TRIMA','KAMA','MAMA','T3','ma_5','ma_30','ma_120','H_line','M_line','L_line','HT','MAVP','MIDPOINT','MIDPRICE','SAR']:
                temp_return = cal_return(temp_data)

            else :
                temp_return = sc.fit_transform(temp_data[:360, np.newaxis]).reshape(360)
#                 temp_return = temp_data[:360]
                
            rx,t,power,scales = generate_wavelet(temp_return,0.1)
            data_scaled = sc.fit_transform(power)
            x_train[count,:,:,j] = data_scaled  
            
        close_mean = data[date_i]['Close'][datetime.time(14,30):datetime.time(14,59)].mean()
        # 涨是1 跌是0
        if data[date_i]['Close'].iloc[360] < close_mean:
                y_train[count] = 1
        else:
                y_train[count] = 0 
        count = count + 1  
       
    print(x_train.shape)
    print(y_train.shape)
    return x_train, y_train




def date_range(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + datetime.timedelta(days = n)
```

## 3.2 Data Distribution

```python
def data_distribution(data, start_date, end_date, full_day, indicator = 'Close'):
    date_time = []
    for date_i in date_range(start_date,end_date):
        date_i = date_i.isoformat()
        if date_i in full_day:
            date_time.append(date_i)
    # print(date_time)
    num = len(date_time)
    y_train = np.zeros([num],dtype = np.float32)
    count = 0
    for date_i in tqdm(date_time):       
        close_mean = data[date_i][indicator][datetime.time(14,30):datetime.time(14,59)].mean()
        y_train[count] = (close_mean - data[date_i][indicator].iloc[359])/data[date_i][indicator].iloc[359]
        count = count + 1  
    return y_train
```

```python
def plot_sample_counts(y_total):
    temp = pd.DataFrame(y_total)
    # print(temp[0])
    count = temp[0].value_counts()
    plt.style.use('ggplot')
    count.plot(kind='bar', title='Count (target)');

    print('Class 0:', count[0])
    print('Class 1:', count[1])
    print('Proportion:', round(count.iloc[0] / count.iloc[1], 2), ': 1')
```
